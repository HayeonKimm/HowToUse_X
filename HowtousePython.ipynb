## 문자열 포맷팅

### %s : 문자열 , %c : 문자한개, %d : 정수 

```python
print("I eat %d apples" %3) # 숫자
print("I eat %s apples" %'three') # 문자열
```

```python
# 숫자 값을 나타내는 변수로 대입
number = 3 
'I eat %d apples' %number
```

```python
# 2개 이상의 값 넣기

number = 10 
day = 'three'
'I ate %d apples. so I was sick for %s' %(number,day)
```
